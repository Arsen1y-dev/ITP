Генератор матриц

```python
def make_matrix(size, value=0):
    if isinstance(size, int):
        width = size
        height = size
    else:
        width, height = size

    matrix = [[value for _ in range(width)] for _ in range(height)]
    return matrix
```

Код make_matrix работает по принципу построения матрицы с помощью вложенных списков и циклов.
1. Определение размера матрицы:
Функция сначала определяет, какой размер матрицы требуется:
Если size - целое число, это означает, что нужна квадратная матрица, поэтому ширина и высота устанавливаются равными этому числу.
Если size - кортеж, это означает, что требуется матрица с различной шириной и высотой, и эти значения извлекаются из кортежа.
2. Создание пустой матрицы:
Функция создает пустую матрицу (список списков), используя вложенные списковые выражения:
[[value for _ in range(width)] for _ in range(height)]
Внешний цикл (for _ in range(height)) проделывает следующее:
Создает пустой список (будущую строку матрицы).
Запускает внутренний цикл.
Внутренний цикл ([value for _ in range(width)]) заполняет созданный список:
Добавляет значение value в список width раз.
Этот список является строкой матрицы.
3. Возвращение матрицы:
После того, как матрица создана, функция возвращает ее как результат.
Пример:
Допустим, size = (3, 2) и value = 5:
Определение размера: width = 3, height = 2.
Создание матрицы:
Создается первый список (первая строка матрицы). В цикле добавляется value (5) 3 раза: [5, 5, 5].
Создается второй список (вторая строка матрицы). В цикле добавляется value (5) 3 раза: [5, 5, 5].
Матрица теперь: [[5, 5, 5], [5, 5, 5]].
Возврат матрицы: Функция возвращает матрицу [[5, 5, 5], [5, 5, 5]].
Ключевые моменты:
Функция make_matrix использует вложенные списки для представления матрицы.
Вложенные списковые выражения обеспечивают компактный способ создания и заполнения матрицы.
Функция возвращает матрицу, которую можно использовать для дальнейших операций.

Функциональный нод 2.0

```python
def gcd(*numbers):
    if len(numbers) == 1:
        return numbers[0]
    else:
        result = numbers[0]
        for i in range(1, len(numbers)):
            result = _gcd(result, numbers[i])
        return result


def _gcd(a, b):
    while b:
        a, b = b, a % b
    return a
```

Код реализует алгоритм Евклида для нахождения наибольшего общего делителя (НОД). Давайте разберем его по частям:
1. Функция _gcd(a, b):
Эта вспомогательная функция вычисляет НОД двух чисел a и b.
Она использует цикл while b:, который выполняется, пока b не станет равно 0.
Внутри цикла происходит следующее:
a, b = b, a % b: Здесь происходит ключевой шаг алгоритма Евклида. a заменяется на b, а b заменяется на остаток от деления a на b.
Когда b становится равным 0, цикл завершается, и функция возвращает текущее значение a, которое и является НОД.
2. Функция gcd(*numbers):
Эта функция принимает произвольное количество чисел *numbers.
if len(numbers) == 1:: Если передан только один аргумент, то он и является НОД.
else:: Если передано несколько чисел, то выполняется следующий алгоритм:
result = numbers[0]: Начальное значение НОД устанавливается равным первому числу в последовательности.
for i in range(1, len(numbers)):: Цикл проходит по всем оставшимся числам.
result = _gcd(result, numbers[i]): На каждом шаге вызывается _gcd с текущим значением НОД (result) и следующим числом в последовательности (numbers[i]). Результат этой операции становится новым значением result.
return result: После обработки всех чисел, функция возвращает окончательное значение result, которое и является НОД всей последовательности.
Принцип работы алгоритма Евклида:
Алгоритм Евклида основан на следующем свойстве: НОД двух чисел равен НОД меньшего из них и остатка от деления большего на меньшее.
Повторяя этот процесс, мы в конечном итоге придем к ситуации, когда остаток от деления станет равен 0. В этот момент меньшее число и будет НОД исходных чисел.
Пример:
Найдем НОД чисел 36 и 48:
gcd(36, 48)
_gcd(36, 48)
a = 48, b = 36 % 48 = 36
a = 36, b = 48 % 36 = 12
a = 12, b = 36 % 12 = 0
b = 0, возвращаем a = 12.
Таким образом, gcd(36, 48) = 12.

Подготовка данных

```python
def to_string(*data, sep=" ", end="\n"):
    string = sep.join(str(item) for item in data)
    return string + end
```

Определение функции:
def to_string(*data, sep=" ", end="\n"):
Функция to_string принимает неопределенное количество аргументов *data (это обозначает, что функция может принимать любое количество аргументов, как один, так и множество).
Она также имеет два необязательных параметра: sep (разделитель) и end (символ в конце строки). По умолчанию sep - это пробел, а end - символ новой строки.
Объединение данных в строку:
string = sep.join(str(item) for item in data)
str(item) for item in data: Создается генератор, который преобразует каждый элемент data в строку. Это необходимо, потому что sep.join работает только с последовательностями строк.
sep.join(...): Метод join объединяет элементы из генератора, вставляя между ними sep.
Добавление символа конца строки:
return string + end
string + end: К полученной строке добавляется символ end.
Пример:
result = to_string(1, 2, 3)
print(result)  # Вывод: 1 2 3\n
to_string(1, 2, 3) - Вызов функции с тремя числами.
sep по умолчанию - пробел, поэтому числа будут разделены пробелами.
end по умолчанию - символ новой строки.
Другой пример:
data = [7, 3, 1, "hello", (1, 2, 3)]
result = to_string(*data, sep=", ", end="!")
print(result)  # Вывод: 7, 3, 1, hello, (1, 2, 3)!
*data: Распаковывает список data в отдельные аргументы для функции to_string.
sep=", ": Устанавливает разделитель в виде запятой с пробелом.
end="!": Устанавливает символ конца строки в виде восклицательного знака.

```python
def order(*args):
    temp = in_stock
    GRADES = {
        "Эспрессо": {"coffee": 1},
        "Капучино": {"coffee": 1,
                     "milk": 3},
        "Макиато": {"coffee": 2,
                    "milk": 1},
        "Кофе по-венски": {"coffee": 1,
                           "cream": 2},
        "Латте Макиато": {"coffee": 1,
                          "milk": 2,
                          "cream": 1},
        "Кон Панна": {"coffee": 1,
                      "cream": 1},
    }
    for grade in args:
        for ingr in GRADES[grade]:
            if GRADES[grade].get(ingr, 0) > in_stock[ingr]:
                break
        else:
            for ingr in GRADES[grade]:
                in_stock[ingr] -= GRADES[grade][ingr]
            return grade
    if in_stock == temp:
        return "К сожалению, не можем предложить Вам напиток"
```

Функция order принимает произвольное количество аргументов (названия напитков).
Внутри функции создается локальная переменная temp, которая ссылается на текущий остаток ингредиентов (in_stock).
Словарь GRADES содержит рецепты напитков, где ключи — названия напитков, а значения — словари с необходимыми ингредиентами и их количествами.
Внешний цикл for grade in args: проходит по каждому переданному названию напитка.
Вложенный цикл for ingr in GRADES[grade]: проверяет, достаточно ли ингредиентов для приготовления напитка. 
Если какого-либо ингредиента не хватает, цикл прерывается и переходит к следующему напитку.
Если всех ингредиентов достаточно, они вычитаются из in_stock, и функция возвращает название напитка.
Если ни один напиток не может быть приготовлен, функция возвращает сообщение "К сожалению, не можем предложить Вам напиток".

```python
numbers = tuple()


def enter_results(*args):
    global numbers
    numbers += args


def get_sum():
    return round(sum(numbers[::2]), 2), round(sum(numbers[1::2]), 2)


def get_average():
    return round(2 * get_sum()[0] / len(numbers), 2), round(2 * get_sum()[1] / len(numbers), 2)
```

numbers — глобальная переменная, хранящая кортеж чисел.
Функция enter_results:Принимает произвольное количество аргументов и добавляет их к глобальной переменной numbers.
Функция get_sum:Возвращает кортеж из двух значений:Суммы чисел на четных позициях в numbers.
Суммы чисел на нечетных позициях в numbers.
Использует срезы numbers[::2] (четные позиции) и numbers[1::2] (нечетные позиции) и округляет результат до двух знаков после запятой.Функция get_average:Возвращает кортеж из двух значений:Среднее значение чисел на четных позициях.
Среднее значение чисел на нечетных позициях.
Использует результаты функции get_sum и делит их на длину кортежа numbers, умножая на 2, чтобы учесть оба значения (суммы) и получить среднее значение для четных и нечетных позиций отдельно.

Ключевой секрет

```python
def secret_replace(text, **replaces):
    result = []
    for char in text:
        if char in replaces:
            result.append(replaces[char][0])
            replaces[char] = replaces[char][1:] + replaces[char][:1]
        else:
            result.append(char)
    return ''.join(result)
```

Проход по символам текста: Мы проходим по каждому символу строки text.Проверка замен: Если текущий символ char есть в словаре replaces, мы заменяем его на первое значение из кортежа замен.
Обновление замен: После замены, чтобы использованное значение снова не использовалось первым, мы перемещаем его в конец кортежа замен.Сбор результата: Мы собираем измененные символы в список result, который затем объединяется в итоговую строку.
Эта реализация обеспечивает циклическое использование замен, что соответствует правилам, заданным в условии задачи.
