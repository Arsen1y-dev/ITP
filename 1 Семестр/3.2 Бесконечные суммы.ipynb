3.2 Бесконечные суммы

Пример

```python
h, e = map(float, input().split())

x = 0.2

while 1.8 + e > x:
    y = x
    y_sum = 0

    while e < y:
        y_sum += y
        y = (1/13) * (x * y ** 2 + 2 * y)

    print(f'{x:.8f}', f'{y_sum:.8f}')
    x += h
```

```python
h, e = map(float, input().split())

x = 0.2

while 0.8 + e > x:
    y0, y1 = 1, x
    i = 1
    y_sum = y0 + y1

    while e < y1:
        try:
            y0, y1 = y1, ((((x ** 2 - 2) * y1) / ((x ** 2 - 1) * i ** 2)) + (y0 ** 2 / i ** 2))
            y_sum += y1
            i += 1

        except OverflowError:
            y_sum = 0
            break

    print(f'{x:.8f}', f'{y_sum:.8f}')
    x += h
```

```python
h, e = map(float, input().split())

x = 0.1

while 0.9 + e > x:
    y = 1 + x / 4
    i = 0
    i1 = 1
    y_sum = 0

    while e < y:
        y_sum += (-1) ** i * y
        i += 1
        i1 += 1
        y = x ** i1 / (i1 * (i1 + 3))

    print(f'{x:.8f}', f'{y_sum:.8f}')
    x += h
```
