Задание 4.1 Обработчик строк

```python
string = input('Введите строку:')
l = 0
while True:
    print('Выберите действие:')
    print('1. Показать текущую строку в двойных кавычках и в скобках её длину.')
    print('2. Заменить текущую строку, прочитав новую.')
    print('3. Повторить строку n раз и дополнить справа до заданной длины определённым символом')
    print('4. Проверить, встречается ли одна подстрока раньше другой')
    print('5. Подсчитать количество символов текущей строки, не встречающихся во введённой строке, без учёта регистра')
    print('6. Завершить работу.')

    choice = int(input("Введите номер действия от 1 до 6: "))

    if choice in (1, 2, 3, 4, 5, 6):
        if choice == 1:
            print(f'"{string}" ({len(string)})')
        elif choice == 2:
            string = input('Введите новую строку: ')
            print(string)
        elif choice == 3:
            q = int(input('Выберите колличество раз: '))
            p = int(input('Введите желаемую длину: '))
            r = str(input('Введите желаемый символ: '))
            string = string * (q + 1)
            while p > len(string):
                string = string + (p - len(string)) * r
            print(string)
        elif choice == 4:
            p = str(input("Введите подстроку, которая встречается раньше: "))
            d = str(input("Введите подстроку, которая встречается позже: "))
            index1 = string.find(p)
            index2 = string.find(d)
            if index1 < index2:
                print(True)
            else:
                print(False)
        elif choice == 5:
            p = str(input("Введите подстроку: "))
            for i in string:
                if i.lower() not in p.lower():
                    l += 1
            print(l)
        elif choice == 6:
            print('Работа завершена.')
            break
    else:
        print('Некорректный ввод. Пожалуйста, выберите действие от 1 до 6.')
```

```python
s = input('Введите строку:')
while True:
    print('Выберите действие:')
    print('1. Показать текущую строку в двойных кавычках и в скобках её длину.')
    print('2. Заменить текущую строку, прочитав новую.')
    print('3. Убрать пробельные символы из начала строки и заменить подстроку звёздочками.')
    print('4. Проверить, совпадают ли первая и предпоследняя буквы строки.')
    print('5. Подсчитать количество символов текущей строки, не встречающихся во введённой строке.')
    print('6. Завершить работу.')
    
    choice = int(input("Введите номер действия от 1 до 6: "))
    
    if choice in (1, 2, 3, 4, 5, 6):
        if choice == 1:
            print(f'"{s}" ({len(s)} символов)')
        elif choice == 2:
            s = input('Введите новую строку: ')
            print(s)
        elif choice == 3:
            s = s.lstrip().replace(input('Введите подстроку: '), '***')
            print(s)
        elif choice == 4:
            if s[0] == s[-2]:
                print('Первая и предпоследняя буквы совпадают.')
            else:
                print('Первая и предпоследняя буквы не совпадают.')
        elif choice == 5:
            s1 = input('Введите строку: ')
            diff_chars = [char for char in s if char not in s1]
            print('Количество символов текущей строки, не встречающихся во введённой строке:', len(diff_chars))
        elif choice == 6:
            print('Работа завершена.')
            break
    else:
        print('Некорректный ввод. Пожалуйста, выберите действие от 1 до 6.')
```
