```python
import time

def intro():
    print("Добро пожаловать в детективную игру!")
    print("Вы столкнулись с загадочным преступлением и решаете взять расследование в свои руки.")

def crime_scene():
    print("\nВы находитесь на месте преступления. Перед вами дом, из которого доносится странный свет.")
    print("1. Поговорить со свидетелем")
    print("2. Заглянуть в шкаф")
    print("3. Перейти в другую комнату")
    print("4. Выйти на улицу")

def talk_with_witness():
    print("\nВы разговариваете со свидетелем, который рассказывает о странных звуках в доме.")
    print("1. Вернуться на место преступления")

def look_in_closet(has_flashlight):
    if has_flashlight:
        print("\nВы находите фонарик в шкафу.")
    else:
        print("\nВы заглядываете в шкаф, но ничего не находите.")
    print("1. Вернуться на место преступления")

def choose_room():
    print("\nВы выбираете, в какую комнату перейти.")
    print("1. Зайти в подвал")
    print("2. Вернуться на место преступления")

def basement(has_flashlight):
    if has_flashlight:
        print("\nС фонариком в руках, вы спускаетесь в подвал и обнаруживаете следы.")
        print("1. Обследовать следы")
        print("2. Вернуться на место преступления")
    else:
        print("\nВы не можете зайти в подвал без фонарика.")
        print("1. Вернуться на место преступления")

def investigate_traces():
    print("\nВы обнаруживаете следы, которые ведут к тайной комнате.")
    print("1. Войти в тайную комнату")
    print("2. Вернуться на место преступления")

def secret_room():
    print("\nВ тайной комнате вы находите дополнительные улики и разгадку преступления.")
    print("Поздравляю, вы завершили расследование!")

def accuse():
    print("\nПришло время обвинить кого-то в преступлении.")
    print("1. Обвинить Алису")
    print("2. Обвинить Боба")

def ending(result):
    if result == "win":
        print("\nВы правильно разгадали преступление! Поздравляю, вы выиграли.")
    else:
        print("\nВаш выбор был неверным. Попробуйте снова.")

# Основной код
intro()
crime_scene()

choice = input("\nВыберите действие: ")

if choice == "1":
    talk_with_witness()
elif choice == "2":
    has_flashlight = False
    look_in_closet(has_flashlight)
    has_flashlight = True  # После нахождения фонарика
elif choice == "3":
    choose_room()
    choice = input("\nВыберите действие: ")
    if choice == "1":
        basement(has_flashlight)
        choice = input("\nВыберите действие: ")
        if choice == "1":
            investigate_traces()
            secret_room()
        elif choice == "2":
            crime_scene()
    elif choice == "2":
        crime_scene()
elif choice == "4":
    crime_scene()

accuse_choice = input("\nХотите обвинить кого-то в преступлении? (y/n): ")

if accuse_choice.lower() == "y":
    accuse()
    accuse_result = input("\nВаш выбор: ")
    if accuse_result == "1":
        ending("win")
    elif accuse_result == "2":
        ending("lose")
else:
    print("\nВы решили не обвинять никого. Игра завершена.")
```

```python
import time

def intro():
    print("Добро пожаловать в детективную историю!")
    print("Вы - детектив, расследующий загадочное убийство.")
    print("Вы находитесь в комнате с трупом. Что вы хотите сделать?")
    time.sleep(1)
    print("1. Поговорить со свидетелем.")
    print("2. Заглянуть в шкаф.")
    print("3. Перейти в другую комнату.")
    print("4. Выйти на улицу.")
    choice = input("Выберите действие (1/2/3/4): ")
    if choice == "1":
        interrogate_witness()
    elif choice == "2":
        check_closet()
    elif choice == "3":
        enter_another_room()
    elif choice == "4":
        go_outside()
    else:
        print("Неправильный выбор. Попробуйте еще раз.")
        intro()

def interrogate_witness():
    print("Вы решили поговорить со свидетелем.")
    print("Свидетель говорит, что видел нечто странное ночью.")
    time.sleep(1)
    print("1. Спросить, что именно он видел.")
    print("2. Спросить, есть ли у него какие-либо улики.")
    print("3. Вернуться назад.")
    choice = input("Выберите действие (1/2/3): ")
    if choice == "1":
        print("Свидетель рассказывает вам о темной фигуре, убегающей с места преступления.")
        time.sleep(1)
        interrogate_witness()
    elif choice == "2":
        print("Свидетель говорит, что нашел нож возле трупа.")
        time.sleep(1)
        interrogate_witness()
    elif choice == "3":
        intro()
    else:
        print("Неправильный выбор. Попробуйте еще раз.")
        interrogate_witness()

def check_closet():
    print("Вы заглянули в шкаф и нашли фонарик.")
    time.sleep(1)
    intro()

def enter_another_room():
    print("Вы перешли в другую комнату, но там ничего нет.")
    time.sleep(1)
    intro()

def go_outside():
    print("Вы вышли на улицу и видите темное небо.")
    print("1. Осмотреться вокруг.")
    print("2. Вернуться назад.")
    choice = input("Выберите действие (1/2): ")
    if choice == "1":
        print("Вы замечаете следы, ведущие в лес.")
        time.sleep(1)
        investigate_forest()
    elif choice == "2":
        intro()
    else:
        print("Неправильный выбор. Попробуйте еще раз.")
        go_outside()

def investigate_forest():
    print("Вы идете в лес, следуя за следами.")
    print("1. Пойти дальше вглубь леса.")
    print("2. Вернуться на улицу.")
    choice = input("Выберите действие (1/2): ")
    if choice == "1":
        print("Вы находите заброшенную хижину и внутри обнаруживаете улику.")
        time.sleep(1)
        accuse_suspects()
    elif choice == "2":
        go_outside()
    else:
        print("Неправильный выбор. Попробуйте еще раз.")
        investigate_forest()

def accuse_suspects():
    print("У вас есть несколько подозреваемых:")
    print("1. Мистер А")
    print("2. Миссис Б")
    print("3. Доктор В")
    choice = input("Кого вы хотите обвинить? (1/2/3): ")
    if choice == "1":
        print("Вы обвинили Мистера А, и это был верный ответ! Вы разгадали дело!")
    elif choice == "2" or choice == "3":
        print("Вы обвинили неверного человека. Дело остается нераскрытым.")
    else:
        print("Неправильный выбор. Попробуйте еще раз.")
        accuse_suspects()

# Начать игру с ввода
intro()
```
